```python
import numpy as np
import matplotlib.pyplot as plt

X_mu = np.array([6, 2, 4, 3, 4, 5, 0, 0, 0, 0], dtype=complex)
#X_mu = np.array([8, 2, 4, 0, 0, 0, 0, 0], dtype=complex)
N = len(X_mu)
```

```python
k = np.arange(N)
mu = np.arange(N)
W = np.exp(1j * 2 * np.pi / N * np.outer(k, mu))
```

```python
x_k = (1/N) * W @ X_mu
```

```python
x_ifft = np.fft.ifft(X_mu)
np.allclose(x_k, x_ifft)
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.stem(k, np.real(x_k))
ax.set_title('Sygnał x[k] z IDFT')
ax.set_xlabel('k (próbka)')
ax.set_ylabel('Amplituda (Re{x[k]})')
ax.grid(True)
plt.show()
```
